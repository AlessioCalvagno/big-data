```
from pyspark.ml.feature import VectorAssembler,StandardScaler,PCA,Imputer
from pyspark.ml import Pipeline
```

```
df = spark.table("dataset")
```

```
train,test = df.randomSplit([0.8,0.2],seed=0)
```

```
train.dtypes
```

```
numerical_variables = [x[0] for x in train.dtypes if x[1] != "string"]
```

```
numerical_variables
```

```
pipe = Pipeline(stages = [
    Imputer(inputCols=numerical_variables,outputCols=numerical_variables),
    VectorAssembler(inputCols=numerical_variables,outputCol="clean_features"),
    PCA(k=5,inputCol="clean_features",outputCol="pca_features"),
    StandardScaler(inputCol="pca_features",outputCol="features")
    
])
```

```
model = pipe.fit(train)
```

```
display(model.transform(train))
```
